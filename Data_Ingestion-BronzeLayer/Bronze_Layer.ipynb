### now we are working with each file to extract the data and metadata.

```python
train_identity_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("/Volumes/ml_team/devesh/raw_data/train_identity.csv")
```

```python
train_identity_df.printSchema()
```

```python
train_identity_df.count()
```

```python
train_identity_df.write.mode("overwrite").saveAsTable("ml_team.devesh.train_identity")
```

```python
%sql
select * from ml_team.devesh.train_identity
limit 10
```
