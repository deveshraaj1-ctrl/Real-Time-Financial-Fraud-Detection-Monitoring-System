```python
# Read Silver tables in BATCH
base_df = spark.read.table(
    "ml_team.devesh_silverlayer.silver_transactions_base"
)

features_df = spark.read.table(
    "ml_team.devesh_silverlayer.silver_transactions_base"
)

# Time-range batch join (SAFE)
training_df = (
    base_df.join(
        features_df,
        (base_df.card1 == features_df.card1) &
        (base_df.event_timestamp >= features_df.window_start) &
        (base_df.event_timestamp < features_df.window_end),
        "left"
    )
)

training_df.show(10)
training_df.count()
```
