### Fraud Metrics Over Time (CORE TABLE)
This table answers - “How many frauds are happening over time?"

```python
%sql
CREATE OR REPLACE TABLE angad_kumar91.fraud_detection_goldlayer.gold_fraud_metrics_time
USING DELTA
AS
SELECT
    date_trunc('minute', event_timestamp) AS minute_ts,
    COUNT(*) AS total_txns,
    SUM(is_anomaly) AS fraud_txns,
    ROUND(SUM(is_anomaly) / COUNT(*), 4) AS fraud_rate
FROM angad_kumar91.fraud_detection_goldlayer.fraud_predictions
GROUP BY date_trunc('minute', event_timestamp)
ORDER BY minute_ts;
```

```python
%sql
SELECT *
FROM angad_kumar91.fraud_detection_goldlayer.gold_fraud_metrics_time
ORDER BY minute_ts DESC;
```

### High-Risk Cards Table

What this table answers -  “Which cards are most risky?”

```python
%sql
CREATE OR REPLACE TABLE angad_kumar91.fraud_detection_goldlayer.gold_high_risk_cards
USING DELTA
AS
SELECT
    card1,
    COUNT(*) AS total_txns,
    SUM(is_anomaly) AS fraud_txns,
    ROUND(SUM(is_anomaly) / COUNT(*), 4) AS fraud_rate
FROM angad_kumar91.fraud_detection_goldlayer.fraud_predictions
GROUP BY card1
HAVING SUM(is_anomaly) > 0
ORDER BY fraud_txns DESC;
```

```python
%sql
SELECT *
FROM angad_kumar91.fraud_detection_goldlayer.gold_high_risk_cards;
```

### Latest Fraud Alerts
What this table answers - “What are the most recent fraud alerts right now?”

```python
%sql
CREATE OR REPLACE TABLE angad_kumar91.fraud_detection_goldlayer.gold_latest_fraud_alerts
USING DELTA
AS
SELECT
    TransactionID,
    event_timestamp,
    card1,
    TransactionAmt,
    fraud_score
FROM angad_kumar91.fraud_detection_goldlayer.fraud_predictions
WHERE is_anomaly = 1
ORDER BY event_timestamp DESC;
```

```python
%sql
SELECT *
FROM angad_kumar91.fraud_detection_goldlayer.gold_latest_fraud_alerts;
```
